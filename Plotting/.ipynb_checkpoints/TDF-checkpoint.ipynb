```python
import numpy as np
import matplotlib.pyplot as plt


en  = np.zeros(13945)
tdf  = np.zeros(13945)

      
with open('tdf.dat') as fp:
    for i, line in enumerate(fp):
        item = line.split()   
        en[i] = float(item[0]) 
        tdf[i] = float(item[1])       
        
              
#SI units        
# eV to J        
en = (en - 0.1520)*1.60217662e-19


# height of unit cell in A
h = 26.2

# electron charge in Coul
e_charge = 1.60217662e-19

#Plank constant (J/c)
hbar = 1.054571e-34

#relaxation time in fs
tao = 10 

  
omega = 0.0
for i in range(13945):
    omega+= (e_charge**2 * h)/(hbar * tao) *tdf[i]
        
        
energy  = np.zeros(352)

with open('kubo_S_xx.dat') as fp:
    for i, line in enumerate(fp):
        item = line.split()     
        energy[i] = float(item[0])
        
   
for i in range(352):
    sigma_intra =  1/(0.01 - 1j * energy[i]) * omega
    print(energy[i], '\t', (sigma_intra).real, '\t', (sigma_intra).imag)
              
```
