```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors


def ddf(x, shift, sig):
    val = np.zeros_like(x)
    val[(-(1/(2*sig))<=(x-shift)) & ((x-shift)<=(1/(2*sig)))] = 1
    return val


thao = np.zeros((4096,24))
ebnd = np.zeros((4096,24))

with open('scattering.dat') as fp:
    for line in fp:
        item = line.split()  
        if(line=='\n' or line.strip()=='' or str.startswith(item[0],'#')):
            continue #skip comments and blank lines    
        thao[int(item[0])-1][int(item[1])-1] = float(item[3])
        ebnd[int(item[0])-1][int(item[1])-1] = float(item[2])


dos = np.zeros(907) 
energy = np.zeros(907) 

with open('dos.dat') as fp:
    for i,line in enumerate(fp):
        item = line.split()  
        energy[i] = float(item[0])  
        dos[i] = float(item[1])  

        
# for e in range(1089):
#     print(energy[e], ddf(energy[e], -4, 5))
    
      
               
thao_sum = 0.0
for e in range(907):
    for k in range(4096):
        for n in range(24):
            thao_sum +=thao[k][n]* ddf(energy[e], ebnd[k][n], 5)/dos[e]
    print(e)        
            
thao_sum = thao_sum/4096

print(thao_sum)            
        
  
    
# plt.rcParams["font.family"] = "serif"
# plt.rcParams["font.serif"] = "Times New Roman"


# fig = plt.figure(figsize=(8,8))
# ax = fig.add_subplot(111)
# im = ax.set_aspect(1.5)
# im = ax.plot(t, res_up,  color='Royalblue', label=r'$\rho^\uparrow$(FGT)', linewidth=2)
# im = ax.plot(t, res_dn,  color='magenta', label=r'$\Delta \rho$ (FGT)', linewidth=2)
# im = ax.plot(t, res_dn,  color='Red', label=r'$\rho^\downarrow$(FGT)', linewidth=2)
# im = ax.plot(t_fe, rho_fe,  color='green', label=r'$\rho$(Fe)', linewidth=1, marker='o',markersize=5)
# im = ax.plot(t_fe, rho_cu,  color='orange', label=r'$\rho$(Cu)', linewidth=1, marker='o',markersize=5)
# im = ax.grid(color='black', linestyle='dashed', linewidth=0.5)
# ax.set_xlim([0, 400])
# ax.set_ylim([0, 100])
# ax.set_ylabel('Resistivity', fontsize=18)
# ax.set_xlabel('Temperature, T (K)', fontsize=18)
# ax.tick_params(axis = 'both', which = 'major', labelsize = 18)
# ax.legend(loc='best', fontsize=18)

# ax2 = ax.twinx()
# # ax2.set_aspect(1.5)
# ax2.plot(t, res_spin, color='magenta', linewidth=2 )
# ax2.tick_params(axis = 'both', which = 'major', labelsize = 18)
# ax2.set_ylabel(r'$\Delta \rho$', fontsize=18)


# fig.savefig('dn.png', dpi=300)




    
    

```
