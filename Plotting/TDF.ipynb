```python
import numpy as np
import matplotlib.pyplot as plt


tdf  = np.zeros(13945)

kmesh = 100000

      
with open('tdf.dat') as fp:
    for i, line in enumerate(fp):
        item = line.split()   
        tdf[i] = float(item[1]) # (eV fs/Ang) (1/hbar^2 )    
        
                     
# height of unit cell in A
h = 26.2

# electron charge in Coul
e_charge = 1.60217662e-19

#Plank constant (Jc)
hbar = 1.054571e-34


#relaxation time in fs
tao = 10 
  
omega = 0.0
for i in range(13945):
    omega+= (e_charge**2 * h)/(tao * hbar * kmesh) * tdf[i] * 1.60217662e-19
        
        
energy  = np.zeros(352)

with open('kubo_S_xx.dat') as fp:
    for i, line in enumerate(fp):
        item = line.split()     
        energy[i] = float(item[0])
        
   
for i in range(352):
    sigma_intra =  1/(0.01*1.60217662e-19 - 1j * energy[i] * 1.60217662e-19) * omega
    print(energy[i], '\t', (sigma_intra).real, '\t', (sigma_intra).imag)
              
```
