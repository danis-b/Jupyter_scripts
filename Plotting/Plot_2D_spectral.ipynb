```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors



bands_up = np.zeros((10,383))
# bands_dn = np.zeros((10,383))
Ef = -2.8117
i = 0
j = 0
with open('bands_dn.dat') as fp:
    for line in fp:
        j = i//383
        item = line.split()  
        if(line=='\n' or line.strip()=='' or str.startswith(item[0],'#')):
            continue #skip comments and blank lines    
            print(i, j,'blank line!')
        bands_up[j][i - 383*j] = float(item[1])-Ef
        i+=1
        
# i = 0
# j = 0
# with open('bands_dn.dat') as fp:
#     for line in fp:
#         j = i//383
#         item = line.split()  
#         if(line=='\n' or line.strip()=='' or str.startswith(item[0],'#')):
#             continue #skip comments and blank lines    
#             print(i, j,'blank line!')
#         bands_dn[j][i - 383*j] = float(item[1])-Ef
#         i+=1
        


x = np.arange(383)
y = np.linspace(-0.5, 3.0, 800)
data_up = []
# data_dn = []

with open('dn.dat') as fp:
    for line in fp:
        item = line.split() 
        if(line=='\n' or line.strip()=='' or str.startswith(item[0],'#')):
            continue #skip comments and blank lines 
        data_up =  np.append(data_up, float(item[2]))        

print(max(data_up))        
        
# with open('dn.dat') as fp:
#     for line in fp:
#         item = line.split() 
#         if(line=='\n' or line.strip()=='' or str.startswith(item[0],'#')):
#             continue #skip comments and blank lines 
#         data_dn =  np.append(data_dn, float(item[2]))   
        
                
data2D_up  = data_up.reshape(383, 800).T
# data2D_dn  = data_dn.reshape(383, 800).T

xmesh, ymesh = np.meshgrid(x,y)

print('data scannig completed!')
```
