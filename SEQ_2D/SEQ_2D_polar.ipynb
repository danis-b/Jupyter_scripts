# SEQ 2D_polar

Schrödinger equation is given as:

$$
(-\frac{\hbar^2}{2m} \Delta  - \frac{Ze^2}{r - r_d}) \Psi = E \Psi(r,\phi) 
$$

Laplas operator in polar coordinates

$$
\Delta = \frac{\partial^2}{\partial r^2} + \frac{1}{r} \frac{\partial}{\partial r} + \frac{1}{r^2} \frac{\partial^2}{\partial \phi^2} = \Delta_r + \Delta_\phi   \\
\frac{\partial f}{\partial x} = \frac{f(x+1) - f(x-1)}{2\Delta x} \\
\frac{\partial^2 f}{\partial x^2} = \frac{f(x+1) - 2f(x) + f(x-1)}{\Delta x^2} \\
\\
\\
$$

$$
\Delta_\phi \Psi(x)  =   \frac{\Psi(x+1) - 2\Psi(x) + \Psi(x-1)}{\Delta \phi^2 r^2} \\
\Delta_r \Psi(x) =  \Psi(x+1)[\frac{1}{\Delta r^2} + \frac{1}{2\Delta r r} ] + \Psi(x-1)[\frac{1}{\Delta r^2} - \frac{1}{2\Delta r r} ] - \Psi(x) \frac{2}{\Delta r^2}
$$

Let's consider (3x3) case. Wave function will have the form:
$$
\begin{pmatrix}
\Psi_{r_1, \phi_1} & \Psi_{r_1, \phi_2} & \Psi_{r_1, \phi_3}  \\
\Psi_{r_2, \phi_1} & \Psi_{r_2, \phi_2} & \Psi_{r_2, \phi_3}  \\
\Psi_{r_3, \phi_1} & \Psi_{r_3, \phi_2} & \Psi_{r_3, \phi_3}  \\
\end{pmatrix}
$$

We will present it like vector:
$$
\begin{pmatrix}
\Psi_{r_1, \phi_1} & \Psi_{r_1, \phi_2} &  \Psi_{r_1, \phi_3} & \Psi_{r_2, \phi_1} & ... & \Psi_{r_3, \phi_3}\\ 
\end{pmatrix}^T
$$

Within this basis angular operator $\Delta_\phi$ will have the followig form:
$$
\Delta_\phi = 
\begin{pmatrix}
-\frac{2}{r_1^2 \Delta \phi^2} & \frac{1}{r_1^2 \Delta \phi^2} & 0 & 0 & 0 & 0 & 0 & 0 & 0 \\
\frac{1}{r_1^2 \Delta \phi^2}  & -\frac{2}{r_1^2 \Delta \phi^2} & \frac{1}{r_1^2 \Delta \phi^2} & 0 & 0 & 0 & 0 & 0 & 0 \\
0 & \frac{1}{r_1^2 \Delta \phi^2} & -\frac{2}{r_1^2 \Delta \phi^2} & 0 & 0 & 0 & 0 & 0 & 0 \\
0 & 0 & 0 & -\frac{2}{r_2^2 \Delta \phi^2} & \frac{1}{r_2^2 \Delta \phi^2} & 0 & 0 & 0 & 0 \\
0 & 0 & 0 &\frac{1}{r_2^2 \Delta \phi^2}  & -\frac{2}{r_2^2 \Delta \phi^2} & \frac{1}{r_2^2 \Delta \phi^2} & 0 & 0 & 0 \\
0 & 0 & 0 & 0 & \frac{1}{r_2^2 \Delta \phi^2} & -\frac{2}{r_2^2 \Delta \phi^2} & 0 & 0 & 0 \\
0 & 0 & 0 & 0 & 0 & 0 & -\frac{2}{r_3^2 \Delta \phi^2} & \frac{1}{r_3^2 \Delta \phi^2} & 0 \\
0 & 0 & 0 & 0 & 0 & 0 & \frac{1}{r_3^2 \Delta \phi^2}  & -\frac{2}{r_3^2 \Delta \phi^2} & \frac{1}{r_3^2 \Delta \phi^2} \\
0 & 0 & 0 & 0 & 0 & 0 & 0 & \frac{1}{r_3^2 \Delta \phi^2} & -\frac{2}{r_3^2 \Delta \phi^2} \\
\end{pmatrix}  = D_\phi \otimes \hat{I}
\\
$$ 

Radial operator $\Delta_r$ is
$$
\Delta_r = 
\begin{pmatrix}
-\frac{2}{\Delta r^2} & 0 & 0 & \frac{1}{\Delta r^2}  + \frac{1}{2\Delta r r_2} & 0 & 0 &   0 & 0 & 0 \\
0 &  -\frac{2}{\Delta r^2} & 0 & 0 & \frac{1}{\Delta r^2}  + \frac{1}{2\Delta r r_2} &  0 & 0 & 0 & 0  \\
0 &  0 & -\frac{2}{\Delta r^2} & 0 & 0 & \frac{1}{\Delta r^2}  + \frac{1}{2\Delta r r_2} & 0 & 0 & 0  \\
\frac{1}{\Delta r^2}  - \frac{1}{2\Delta r r_1} & 0 & 0 &  -\frac{2}{\Delta r^2} & 0 & 0 & \frac{1}{\Delta r^2}  + \frac{1}{2\Delta r r_3} & 0 & 0 \\
0 & \frac{1}{\Delta r^2}  - \frac{1}{2\Delta r r_1} & 0 & 0 &  -\frac{2}{\Delta r^2}  & 0 & 0 & \frac{1}{\Delta r^2}  + \frac{1}{2\Delta r r_3} &   0  \\
0 & 0 & \frac{1}{\Delta r^2}  - \frac{1}{2\Delta r r_1} & 0 &  0 & -\frac{2}{\Delta r^2} & 0 & 0 & \frac{1}{\Delta r^2}  + \frac{1}{2\Delta r r_3} \\
0 & 0 & 0 & \frac{1}{\Delta r^2}  - \frac{1}{2\Delta r r_2} & 0 & 0  &  -\frac{2}{\Delta x^2} & 0 & 0  \\
0 & 0 & 0 & 0 & \frac{1}{\Delta r^2}  - \frac{1}{2\Delta r r_2} & 0 & 0 & -\frac{2}{\Delta r^2} &  0 \\
0 & 0 & 0 & 0 & 0 & \frac{1}{\Delta r^2}  - \frac{1}{2\Delta r r_2} &   0 & 0 & -\frac{2}{\Delta x^2} \\
\end{pmatrix} =  \\
 =\hat{I} \otimes D_r
\\
\\
$$

The final Schrödinger equation takes the form:

$$
(-\frac{1}{2m} D_\phi \otimes D_r  - \frac{Ze^2}{r - r_d}) \Psi = E \Psi(r,\phi) 
$$

```python
import numpy as np
import math 
from scipy.sparse.linalg import eigsh
from scipy.sparse.linalg import eigs
from scipy import sparse
import matplotlib.pylab as plt


def get_potential(r, phi, Zval):
#     V = 1*(Zval/(r  + 0.1)) # we multiply to 14.3948  to convert to eV
    V = r*0
    return V


N = 30
Zval = 1 # effective charge
m = 0.02   # effective mass in eV


r = np.linspace(0.1, 10, N, dtype=float)
# r = np.logspace(0, 1, N, dtype=float)
phi = np.linspace(0, 2*np.pi, N, dtype=float) 

r_mesh, phi_mesh = np.meshgrid(r,phi)


dr = 10/N
dp = 2*np.pi/N


# r_mesh = r_mesh.T

V  = get_potential(r_mesh.T, phi_mesh, Zval)


# plt.subplot(projection="polar")
# im = plt.pcolormesh(phi, r, V, shading='nearest')
# plt.colorbar(im)


diag = np.ones([N])
diags_r = np.array([diag*(1/(dr**2) - 1/(2*r*dr) ), -2*diag/(dr**2), diag*(1/(dr**2) + 1/(2*r*dr) )])
D_r = sparse.spdiags(diags_r, np.array([-1,0,1]),N,N)


diags_phi = np.array([diag/(r*dp), -2*diag/(r*dp), diag/(r*dp)])

diags_phi[0][N-1] = diags_phi[0][0]
diags_phi[1][N-1] = diags_phi[1][0]
diags_phi[2][N-1] = diags_phi[2][0]

D_phi = sparse.spdiags(diags_phi, np.array([-1,0,1]),N,N)


T = -1/(2*m)*sparse.kronsum(D_phi,D_r)

U = sparse.diags(V.reshape(N**2), (0))
H = T + U

print(H.todense())

print('hamiltonian was constructed!')
```

```python
#eigenproblem

# eigenvalues, eigenvectors = eigsh(H, k = 5, which = 'SM')

Ham = H.todense()

eigenvalues, eigenvectors = np.linalg.eig(Ham)
eigenvectors = np.array(eigenvectors)

idx = eigenvalues.argsort()[::1]   
eigenvalues = np.real(eigenvalues)[idx]
eigenvectors = np.real(eigenvectors)[:,idx]


print('eigenproblem was solved!')
```

```python
#plotting
m = 0

def get_e(n):
    return eigenvectors.T[n].reshape((N,N))

print(eigenvalues[m])

r_mesh = r_mesh.T

plt.figure(figsize=(8,8))
plt.subplot(projection="polar")
plt.pcolormesh(phi, r,  get_e(m)**2, shading='nearest')
```
