```python
import numpy as np
import matplotlib.pylab as plt


def get_potential_ring(x, y, d, a, delta):
    r = np.sqrt(x**2 + y**2)
    V = (-6.4/(2*np.pi*d))/(np.abs(r - d) +delta)*np.exp(-np.abs(r - d)/a) 
    return V


def get_potential_points(x, y, d, a, delta):
    coord = np.array([[-d, 0], [d, 0], [0, -d], [0, d], [d/np.sqrt(2), d/np.sqrt(2)], [d/np.sqrt(2), -d/np.sqrt(2)], [-d/np.sqrt(2), d/np.sqrt(2)], [-d/np.sqrt(2), -d/np.sqrt(2)]])
    V = 0
    for i in range(8):
        r_i = np.sqrt( (x-coord[i][0])**2 + (y-coord[i][1])**2)
        V += (-0.8)/(r_i +delta)*np.exp(-r_i/a)
    
    return V

d  = 5
a  = 0.5
delta = 1
rMax = 10
N = 200

X,Y = np.meshgrid(np.linspace(-rMax, rMax, N, dtype=float), np.linspace(-rMax, rMax, N, dtype=float))

# x = np.linspace(-10, 10, N)

# fig = plt.figure(figsize=(4,4))
# ax = fig.add_subplot(111)
# ax.plot(x, get_potential_ring(x,0, d, a, delta), label='ring')
# ax.plot(x, get_potential_points(x,0, d, a, delta), label='points')
# ax.legend(loc='lower right', fontsize=12)
# fig.savefig('potential_along_line.png', dpi=300)


V_ring  = get_potential_ring(X,Y, d, a, delta)
V_points  = get_potential_points(X,Y, d, a, delta)


fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.set_aspect(1) 
im = ax.pcolor(X,Y,V_ring,   shading='nearest')
plt.colorbar(im)
fig.savefig('ring.png', dpi=300)


fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.set_aspect(1) 
im = ax.pcolor(X,Y,V_points,   shading='nearest')
plt.colorbar(im)
fig.savefig('points.png', dpi=300)
```
