```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors



x = np.zeros(4000)
dos_Ge = np.zeros((2,4000))
dos_Bi = np.zeros((2,4000))



for i in range(1,25):
    with open(f'dosp.{i}.dat') as fp:
        for j,line in enumerate(fp):
            item = line.split()
            x[j] = float(item[0])
            if (i < 6):
                dos_Ge[0][j] +=(float(item[3]) + float(item[5]) + float(item[7]))
                dos_Ge[1][j] -=(float(item[4]) + float(item[6]) + float(item[8]))
            else:
                dos_Bi[0][j] +=(float(item[3]) + float(item[5]) + float(item[7]))
                dos_Bi[1][j] -=(float(item[4]) + float(item[6]) + float(item[8]))

        
            
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.set_aspect(0.2)
ax.set_xlim([-2, 3.0])
plt.xticks([-2, -1.5, -1, -0.5, 0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0])
ax.set_ylim([-7.5, 7.5])
ax.grid(color='black', linestyle='dashed', linewidth=0.5)
ax.plot(x,dos_Ge[0], color='red', linewidth=2, label='Ge($p$)')
ax.plot(x,dos_Ge[1], color='red', linewidth=2)
ax.plot(x,dos_Bi[0], color='royalblue', linewidth=2, label='Bi($p$)')
ax.plot(x,dos_Bi[1], color='royalblue', linewidth=2)
ax.legend()
ax.set_xlabel('Energy (eV)')
ax.set_ylabel('DOS (states/eV)')
plt.title('Ge/Bi/2Au(111)')

fig.savefig('dos.png', dpi=300)        

```
