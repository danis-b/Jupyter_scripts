```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors



x = np.zeros(4000)
dos_Bi = np.zeros(4000)
dos_Bi_pz = np.zeros(4000)



for i in range(1,18):
    with open(f'dosp.{i}.dat') as fp:
        for j,line in enumerate(fp):
            item = line.split()
            x[j] = float(item[0])
            dos_Bi[j]    += (float(item[2]) + float(item[3]) + float(item[4]))
            dos_Bi_pz[j] += float(item[3])
            
            

                 

with open('data.dat', 'a') as fp1:
    for i in range(4000):
        print(x[i],'\t' ,"{:.4f}".format(dos_Bi[i]) , '\t' , "{:.4f}".format(dos_Bi_pz[i]), file=fp1) 
        
            
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.set_aspect(0.07)
ax.set_xlim([-1, 1])
ax.set_ylim([0, 15])
ax.tick_params(axis='both', which='major', labelsize=15)
ax.grid(color='black', linestyle='dashed', linewidth=0.5)
ax.plot(x,dos_Bi_pz, color='red', linewidth=2, label='Bi($p_z$)')
ax.plot(x,dos_Bi, color='royalblue', linewidth=2, label='Bi($p$)')
ax.legend()
ax.set_xlabel('Energy (eV)', fontsize=15)
ax.set_ylabel('DOS (states/eV)', fontsize=15)
plt.title('Bi/2Au(111)', fontsize=15)

fig.savefig('dos.png', dpi=200)  

print('JOB DONE!')

```
