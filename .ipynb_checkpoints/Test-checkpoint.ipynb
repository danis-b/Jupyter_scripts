```python
    uij = np.random.rand(5,5)
    evals, evecs = np.linalg.eig(uij)
    
    uij_conv = np.matmul(np.matmul(np.linalg.inv(evecs.T), np.multiply(np.identity(5),evals)), evecs.T).T
    
    print('Eigenvalues')
    print(evals)
    
    print('\n Eigenvalues')
    print(np.multiply(np.identity(5),evals))
    
    
    
    print('Uij initial')
    for ii in range(0, 5):
        for jj in range(0, 5):
            print(ii, jj, uij[ii,jj])

    print('OK:')
    
    print('Uij just to check')
    for ii in range(0, 5):
        for jj in range(0, 5):
            print(ii, jj, (uij_conv[ii,jj]).real)
```
